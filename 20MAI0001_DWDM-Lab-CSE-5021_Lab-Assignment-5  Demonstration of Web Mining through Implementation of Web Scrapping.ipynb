# A Brief note on Web Scrapping:

In order to perform Web Scrapping : we write code that sends a request to the server that’s hosting the page we specified. 
The server will return the source code — HTML, mostly — for the page (or pages) we requested.
Unlike a web browser, in web scraping code won't interpret the page's source code and display the page visually, instead,  some custom code has to be implemented that filters through the page's source code looking for specific elements we’ve specified, and extracting whatever content we’ve instructed it to extract.

As an example, In order to get  all of the data from inside a table that was displayed on a web page, our code would be written to go through these steps in sequence:
    1)Request the content (source code) of a specific URL from the server
    
    2)Download the content that is returned
   
    3)Identify the elements of the page that are part of the table we want
  
    4)Extract and (if necessary) reformat those elements into a dataset we can analyze or use in whatever way we require.

# A short description on the Components of a Web Page :

While visiting a web page, our web browser makes a request to a web server. 

This request is called a GET request, since we request to get files from the server. 

The server then sends back files that tell our browser how to render the page for us. 

These files typically include:
    
    HTML — the main content of the page.
    
    CSS — used to add styling to make the page look nicer.
    
    JS — Javascript files add interactivity to web pages.
    
    Images — image formats, such as JPG and PNG, allow web pages to show pictures.
   
   After our browser receives all the files, it renders the page and displays it to us. 

# A brief note on HTML :

While performing Web Scraping, we’re interested in the main content of the web page, so we look primarily at the HTML.

HyperText Markup Language (HTML) is the language that web pages are created in. HTML is a markup language that tells a browser how to display content.

HTML consists of elements called tags, most commonly used HTML tags are : 
    
    The most basic tag is the <html> tag. This tag tells the web browser that everything inside of it is HTML.The head tag contains data about the title of the page.
    
    Body Tag - The main content of the web page goes into the body tag.
    
    p tag - defines a paragraph, and any text inside the tag is shown as a separate paragraph.
    
    div tag — indicates a division, or area, of the page.
    
    b tag — bolds any text inside.
    
    i tag — italicizes any text inside.
    
    table tag — creates a table.
    
    form tag — creates an input form.
    
Some of the tags have commonly used names that depend on their position in relation to other tags:

    child — a child is a tag inside another tag. So the two p tags above       are both children of the body tag.
    
    parent — a parent is the tag another tag is inside. Above, the html       tag is the parent of the body tag.
    
    sibiling — a sibiling is a tag that is nested inside the same parent as another tag. 
    
Like head and body are siblings, since they’re both inside html 
and p tags are siblings, since they’re both inside body.

# A brief on the python 'BeautifulSoup library' :

Beautiful Soup is a Python library for extracting structured data from a website. 

It allows us to parse data from HTML and XML files. 

It acts as a helper module and interacts with HTML in a better way than how other available developer tools would interact with a web page.

It works with parsers like lxml and html5lib to provide Python ways of navigating, searching, and modifying the parse tree.

It converts the documents being fetched to Unicode and outgoing documents to UTF-8 unless the document intrinsic doesn't specify an encoding or Beautiful Soup is unable to detect one in the rarest of the cases.

It is also considered to be faster as compared to other general parsing or scraping techniques.
