```python
#Importing necessay Libraries and Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
#Loading our Input Dataset in Jupyter Notebook
data1 = pd.read_csv("store_data.csv", header=None)
print(data1.shape)
```

```python
#Preprocessing Input data
rec = []
for i in range(1, 7501):
    rec.append([str(data1.values[i, j]) for j in range(0, 20)])
```

```python
print(type(rec))
```

```python
#Implementing Association Rule Mining using Apriori Algorithm
as_rul = apriori(rec, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
as_res = list(as_rul)
```
